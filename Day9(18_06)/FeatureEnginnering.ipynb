```python
def compute_gradient(x , y , w , b):
    m,n = x.shape
    dj_dw= np.zeroes(n)
    dj_db = 0 

    for i in range(m):
        err = (np.dot(x[i],w) + b) - y[i]
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err*x[i,j]
            dj_db = dj_db + err

        dj_db = dj_db/m
        dj_dw = dj_dw/m

        return dj_dw , dj_db    
```

```python
def run_gradient_descent_feng(X, y, iterations, alpha):

    m, n = X.shape

    w = np.zeros(n)
    b = 0

    for i in range(iterations):

        dj_dw, dj_db = compute_gradient(X, y, w, b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db

    return w, b
```
