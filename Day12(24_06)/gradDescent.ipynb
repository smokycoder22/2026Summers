```python
def sigmodial(z):
    return 1/(1+np.exp(-z))

def Lossfunc(X ,y , w ,b):
    m = X.shape[0]
    cost = 0
    for i in range (m):
        f_wx = np.dot(w,X[i]) +b
        f_wb = sigmodial(f_wx)
        cost += -y[i]*np.log(f_wb) - (1-y[i])*np.log(1-f_wb)
    return cost
   
```

```python
def compute_grad(X , y , w , b ):
        m,n = X.shape
        dj_dw = np.zeros((n,))                           #(n,)
        dj_db = 0.

        for i in range(m):
            f_wb_i = sigmodial(np.dot(X[i],w) + b)          #(n,)(n,)=scalar
            err_i  = f_wb_i  - y[i]                       #scalar
            for j in range(n):
                dj_dw[j] = dj_dw[j] + err_i * X[i,j]      #scalar
            dj_db = dj_db + err_i
        dj_dw = dj_dw/m                                   #(n,)
        dj_db = dj_db/m                                   #scalar
            
        return dj_db, dj_dw  
```
