```python
def my_dot(a,b):
    x = 0 
    for i in range(a.shape[0]):
        x = x + a[i]*b[i]
    return x    
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([-1, 4, 3, 2])
print(f"my_dot(a, b) = {my_dot(a, b)}")
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([-1, 4, 3, 2])
c = np.dot(a, b)
print(f"NumPy 1-D np.dot(a, b) = {c}, np.dot(a, b).shape = {c.shape} ") 
c = np.dot(b, a)
print(f"NumPy 1-D np.dot(b, a) = {c}, np.dot(a, b).shape = {c.shape} ")
```

```python
np.random.seed(1)
a = np.random.rand(1000)
b = np.random.rand(1000)
t_start = time.time()
c = np.dot(a,b)
t_end = time.time()

print(f"np.dot(a, b) =  {c:.4f}")
print(f"Vectorized version duration: {1000*(t_end-t_start):.4f} ms ")

t_start = time.time()
c = np.dot(a,b)
t_end = time.time()

print(f"my_dot(a, b) =  {c:.4f}")
print(f"loop version duration: {1000*(t_end-t_start):.4f} ms ")
```
