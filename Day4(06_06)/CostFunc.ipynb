```python
def errorSquare(x , y , w , b):
    m = x.shape[0]
    cost_sum = 0
    for i in range(m):
        f_wb = w*x[i] + b 
        sum = f_wb - y[i]
        cost_sum = sum + cost_sum

    total_cost = ((i/2*m)*cost_sum)    
    return total_cost
```
