```python
import copy, math
import numpy as np
import matplotlib.pyplot as plt 
np.set_printoptions(precision=2)
```

we are trying to build a multiple parameter sysstem that consist area of house , No.of rooms , Which floor house is , How old is house .

```python
X_train = np.array([[2104, 14, 1, 5], [1416, 2, 2, 40], [852, 2, 10, 15]])
y_train = np.array([350, 232, 178])
```

x consistes of paramaters and y consist of price in thane in lakhs..

```python
w_init = np.zeros(X_train.shape[1])
b_init = 0. 
```

model prediction with f(w,b) = w(i).x(i) + b where . is dot product 

Now we will predict single loop with multiple parameter

```python
def single_loop_predict(x , w , b):
    n = x.shape[0]
    p = 0
    for i in range(n):
        p_i = x[i]*w[i]
        p = p + p_i
    p = p+b 
    return p
```

```python
def predict(x , w , b):
    p = np.dot(x , w ) +b 
    return p 
```

```python
x_vec = X_train[0,:]
f_wb = predict(x_vec , w_init , b_init)
print(f_wb)
```

```python
def compute_cost(X, y, w, b): 
    
    m = X.shape[0]
    cost = 0.0
    for i in range(m):                                
        f_wb_i = np.dot(X[i], w) + b           #(n,)(n,) = scalar (see np.dot)
        cost = cost + (f_wb_i - y[i])**2       #scalar
    cost = cost / (2 * m)                      #scalar    
    return cost
```

```python
cost = compute_cost(X_train, y_train, w_init, b_init)
print(f'Cost at optimal w : {cost}')
```

```python
def compute_gradient(x , y , w , b):
    m , n = x.shape
    dj_dw = np.zeros((n,))
    dj_db = 0 

    for i in range(m):
        err = (np.dot(x[i],w)+b) - y[i]
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err * x[i, j]
            dj_db = dj_db + err

    dj_dw = dj_dw/m
    dj_db = dj_db/m

    return dj_db, dj_dw
```

```python
dj_db, dj_dw = compute_gradient(X_train, y_train, w_init, b_init)
print(f'dj_db at initial w,b: {dj_db}')
print(f'dj_dw at initial w,b: \n {dj_dw}')
```

```python
def gradient_des(x , y, w_in , b_in , cost_fun , num_iteration , grad_func , alpha):

    j_histroy = []
    w = copy.deepcopy(w_in)  
    b = b_in

    for i in range(num_iteration):
        dj_db ,  dj_dw = grad_func(x , y , w , b)

        w = w - alpha*dj_dw
        b = b - alpha*dj_db

        if i <100000:
            j_histroy.append(cost_fun(x , y , w ,b))

        if i% math.ceil(num_iteration / 10) == 0:
            print(f"Iteration {i:4d}: Cost {j_histroy[-1]:8.2f}   ")
        
    return w, b, j_histroy
```

```python
initial_w = np.zeros_like(w_init)
initial_b = 0.
# some gradient descent settings
iterations = 1000
alpha = 5.0e-7
# run gradient descent 
w_final, b_final, j_history = gradient_des(X_train, y_train, initial_w, initial_b,
                                                    compute_cost, iterations, compute_gradient, 
                                                    alpha)
print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
m,_ = X_train.shape
for i in range(m):
    print(f"prediction: {np.dot(X_train[i], w_final) + b_final:0.2f}, target value: {y_train[i]}")
```
