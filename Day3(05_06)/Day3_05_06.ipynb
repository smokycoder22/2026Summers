```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,2,3 ,4 ])
y_train = np.array([380 , 500 , 550 ,670 ])
print(f"x_train = {x_train}")
```

```python
m = len(x_train)
print(f"{m}")
```

here Marker is what should be marked in graph and c is color of marker

```python
plt.scatter(x_train , y_train , marker ='x' , c ='r')
plt.title("Housing Price")
plt.ylabel("Price in 1000s of dollar")
plt.xlabel("Area ")
plt.show()
```

```python
def model(x,w ,b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b

    return f_wb        
```

plt.legend() -> it is to show the line in graph..

```python
y_out = model(x_train , 5 , 10)
plt.plot(x_train, y_out, marker='o')
plt.title("Housing Price")
plt.ylabel("Price in 1000s of dollar")
plt.xlabel("Area ")
plt.legend()
plt.show()
```
