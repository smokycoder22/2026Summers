```python
def sigmodial(z):
    return 1/(1 + np.exp(-z))

def compute_logistic_cost(x , y , w , b):
    m = x.shape[0]
    cost = 0 

    for i in range(m):
        z = np.dot(w ,x[i]) + b
        f_wb = sigmodial(z)
        cost += -y[i]*np.log(f_wb) - (1-y[i])*np.log(1-f_wb)

    cost = cost/m
    return cost     
```

```python
import numpy as np

def sigmodial(z):
    return 1/(1 + np.exp(-z))

def compute_cost(x , y , w , b):
    m = len(x)

    cost = 0

    for i in range(m):
        f_wb  = w*x[i] + b
        cost += (f_wb - y[i])**2 

    total_cost = (1/(2*m))*cost

    return total_cost
```

```python
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0, 0, 0, 1, 1, 1])

w =  5
b = -3

cost = compute_cost(x_train, y_train, w, b)
print(cost)
```

```python
import numpy as np
import matplotlib.pyplot as plt


def compute_cost(x,y,w,b):
    m = len(x)
    f = w*x + b
    return np.sum((1/2)*(f-y)**2)/(m)

x_train = np.array([0.,1,2,3,4,5])
y_train = np.array([0,0,0,1,1,1])

w_range = np.linspace(-5,5,100)
costs = []

for w in w_range:
    costs.append(compute_cost(x_train,y_train,w,b=-3))

print(costs)
print(w_range)
plt.plot(w_range,costs)
plt.xlabel("w")
plt.ylabel("Cost J(w,b)")
plt.title("Logistic Regression with Squared Error")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0.001, 0.999, 1000)
loss_y1 = -np.log(p)
loss_y0 = -np.log(1 - p)

plt.plot(p, loss_y1, label="y = 1")
plt.plot(p, loss_y0, label="y = 0")

plt.xlabel("Prediction")
plt.ylabel("Loss")
plt.title("Logistic Loss Curves")
plt.legend()
plt.grid(True)
plt.show()
```
