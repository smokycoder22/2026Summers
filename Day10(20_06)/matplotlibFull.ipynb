```python
import matplotlib.pyplot as plt 
```

```python
x = [1, 2 ,3 ,5]
y = [15,20 , 25,60]
y1 = [18,21 , 27,66]
y2 = [19,30 , 35,69]
plt.plot(x,y)
plt.show
```

charachteristic of graph lines

```python
line_Style = dict(marker = ".",
                  markersize = 30 ,
                  linestyle = "solid",
                  linewidth = '4',
                  markeredgecolor = "#1cd3fc"
                  )
plt.plot(x , y , color = "#1c5bfc" , **line_Style)
plt.plot(x , y1 , color ="#1cd3fc" , **line_Style)
plt.plot(x , y2 ,color ="#1cd3fc" , **line_Style)
```

```python
x = np.array([1 ,2 ,3, 4, 5])
figure , axes = plt.subplots(2,2)
axes[0,0].plot(x , x*2 , color = "red",
               marker = "*" 
               )
axes[0,1].plot(x , x*2 , color = "blue",
               marker = "*" 
               )
axes[1,0].plot(x , x*2 , color = "green",
               marker = "*" 
               )
axes[1,1].plot(x , x*2 , color = "yellow",
               marker = "*" 
               )

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
categories = np.array(["Dsa ","WebDev","AImL" , "cybersecurity"])
salary = np.array([10 , 12, 15 , 10])
plt.bar(categories,salary , color="red")
#for horizontal bar chart -> barh()
plt.xlabel("category")
plt.ylabel("salary")
plt.title("JoBS 2026")
plt.show()
```

pie 

```python
categories = np.array(["Dsa ","WebDev","AImL" , "cybersecurity"])
salary = np.array([101 , 121, 115 , 310])
color = ["red" ,"yellow" , "blue" , "pink"]

plt.pie(salary , labels = categories,
                 autopct="%1.1f%%",
                 colors = color,
                 explode=[0,0,0,0.1])
```

```python
x = [ 1,2,3,4,5,6]
y = [12 , 45 ,67,56 , 78 , 90]
y1 = [32 , 48 ,67,55 , 68 , 89]

plt.scatter(x , y , color = "red",
                    alpha = 0.5,#this transparency and 0.5 is 50 percent 
                    s= 200 ,#this is a diameter 
                    label = "ClassA",
                    edgecolors="blue",
                    marker="*"
                            )
plt.scatter(x , y1 , color = "yellow",
                    alpha = 0.5,#this transparency and 0.5 is 50 percent 
                    marker="o")
plt.xlabel("busines")
plt.ylabel("money" \
"")
```
